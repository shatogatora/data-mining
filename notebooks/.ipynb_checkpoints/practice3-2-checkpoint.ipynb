```python
#q2-1
import spacy
from collections import Counter

# spaCyモデルの読み込み
nlp = spacy.load("ja_core_news_sm")  # 日本語モデル

# ファイル読み込み
file_path = "../data/CoARiJ/2014/docs/S1004A4G_business_risks.txt"  # 任意のテキストファイル
with open(file_path, "r", encoding="utf-8") as f:
    text = f.read()

# 形態素解析
doc = nlp(text)
words = [token.text for token in doc if token.pos_ in ["NOUN", "PRON", "PROPN"]]

# 単語の集計
word_counts = Counter(words)
top_5_words = word_counts.most_common(5)

# 結果表示
print(f"総単語数: {sum(word_counts.values())}")
print("頻出単語トップ5:")
for word, count in top_5_words:
    print(f"{word}: {count}回")
```

```python
import os
import MeCab
from collections import Counter

# データセットのディレクトリ
data_dir = "../data/CoARiJ/2014/docs"

# MeCabの準備
mecab = MeCab.Tagger("-Ochasen")

# 単語を集計する関数
def extract_words(text):
    words = []
    node = mecab.parseToNode(text)
    while node:
        # 品詞フィルタリング
        pos = node.feature.split(",")[0]
        if pos in ["名詞"]:  # 名詞だけを対象
            words.append(node.surface)
        node = node.next
    return words

# 全ファイルの処理
all_words = Counter()
for root, _, files in os.walk(data_dir):
    for file in files:
        if file.endswith(".txt"):
            file_path = os.path.join(root, file)
            with open(file_path, "r", encoding="utf-8") as f:
                text = f.read()
                words = extract_words(text)
                all_words.update(words)

# 結果の出力
total_words = sum(all_words.values())
top_20_words = all_words.most_common(20)

print(f"総単語数: {total_words}")
print("頻出単語トップ20:")
for word, count in top_20_words:
    print(f"{word}: {count}回")
```
